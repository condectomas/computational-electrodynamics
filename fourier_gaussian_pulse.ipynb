# Fourier Series Approximation for Gaussian Pulses

This notebook analyzes the convergence of the Fourier Series approximation for an electromagnetic pulse with a Gaussian profile. The objective is to visualize how the sum of discrete harmonic components reconstructs the original signal as the number of terms ($N$) increases.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initial parameters
P = 10
dt = 1
t = np.linspace(-P/2, P/2, 1000)

# N values to observe the series convergence
N_values = [1, 5, 10, 20, 50]

plt.figure(figsize=(10, 6))

# a0 coefficient (constant term)
a0 = dt * np.sqrt(np.pi) / (2 * P)
```

```python
for N in N_values:
    # Create a column vector with all n values from 1 to N
    n = np.arange(1, N + 1)[:, np.newaxis] 
    
    # Calculate all 'an' coefficients simultaneously
    an = (dt * np.sqrt(np.pi) * np.exp(-(np.pi * n * dt / (2 * P))**2) / P)
    
    # Compute the cosine matrix and sum over the 'n' axis in a single step
    cos_terms = an * np.cos(2 * np.pi * t * n / P)
    ft = a0 + np.sum(cos_terms, axis=0)
    
    # Plot the current iteration
    plt.plot(t, ft, label=f'N = {N}')

plt.title('Convergence of Fourier Series for a Gaussian Pulse', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel('Amplitude f(t)', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

#### Conclusions
**Mathematical Convergence:** It is clearly observed that for low values of $N$ (e.g., $N=1, 5$), the approximation is poor and presents a clear broadening. As $N$ increases towards 50, the superposition of cosine waves accurately reproduces the narrow Gaussian envelope of the original pulse.
